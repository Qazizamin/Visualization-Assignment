### Q1) Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

#Set a seed for reproducibility
np.random.seed(0)

# generating synthetic data
x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 1, x.shape)
y = 2 * x + 1 + noise

#Create a DataFrame for Seaborn
data = pd.DataFrame({'X': x, 'Y': y})

#Create a scatter plot with Seaborn
# relplot means relationship plot, by default it gives a scatter plot
sns.relplot(data = data, x = 'X', y = 'Y', color = 'red', marker = 'o')

#Add title and labels
plt.title('Scatter Plot Of Synthetic Data With Seaborn')
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.grid(True)

plt.show()
```

### Q5)  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

#Generate synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility

#Generate random data
n_samples = 100
X = np.random.randn(n_samples, 2)  # Two features

#Create a third feature that is correlated with the first two
Y = X[:, 0] * 0.5 + X[:, 1] * 1.5 + np.random.randn(n_samples) * 0.1  # Adding some noise

#Create a DataFrame
df = pd.DataFrame({
    'Feature1': X[:, 0],
    'Feature2': X[:, 1],
    'Feature3': Y
})

#Compute the correlation matrix
corr_matrix = df.corr()

#Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)

#Add title
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Q1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.graph_objects as go
import plotly.express as px

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

import warnings
warnings.filterwarnings('ignore')

# Prepare the data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                 # Sine of x

# Create a Bokeh plot
output_notebook()  # Use this line if you are running in a Jupyter notebook
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label='sin(x)', line_width=2)

# Show the plot
show(p)
```

### Q2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

# Prepare the data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(5, 50, size=n)  # Random sizes
colors = np.random.choice(['red', 'orange', 'green', 'pink', 'purple'], size=n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a Bokeh scatter plot
output_notebook()  # Use this line if you are running in a Jupyter notebook
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter markers
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```

### Q3)  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Data
fruits = ['Apple', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh bar chart
output_notebook()  # Use this line if you are running in a Jupyter notebook
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Add bars
p.vbar(x=fruits, top=counts, width=0.9, color="skyblue")

# Labels
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Show the plot
show(p)
```

### Q4) Create a Bokeh histogram to visualize the distribution of the given data.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh histogram
output_notebook()  # Use this line if you are running in a Jupyter notebook
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Count',
           tools="")

# Add bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')

# Show the plot
show(p)
```

### Q5) Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Generate random data for the heatmap
data_heatmap = np.random.rand(10, 10)

# Create a Bokeh heatmap
output_notebook()  # Use this line if you are running in a Jupyter notebook
p = figure(title="Heatmap", width=400, height=400)

# Add rectangles for the heatmap
for i in range(data_heatmap.shape[0]):
    for j in range(data_heatmap.shape[1]):
        color = data_heatmap[i, j]
        p.rect(x=j, y=i, width=1, height=1, color=(f"#{int(color * 255):02x}00{int((1-color) * 255):02x}"))

# Show the plot
show(p)
```
